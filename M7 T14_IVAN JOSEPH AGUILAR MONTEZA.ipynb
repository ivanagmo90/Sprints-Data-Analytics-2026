#### 2.La funció ha de retornar un diccionari amb les següents dades relacionades amb el continent indicat:
- Nombre de regions diferents que hi ha dins del continent.
- Nombre de països dins del continent.
- Superfície màxima (SurfaceArea) registrada al continent.
- Superfície mínima registrada.
- Població mitjana (Population) dels països del continent.
- Mitjana de l’esperança de vida (LifeExpectancy) dels països del continent.
#### 3.Finalment, carrega el fitxer adjunt amb la següent instrucció:
- data = pd.read_csv("arxiu csv", sep=";", encoding="latin-1") Ejecutar antes #Cargar DataFrame

Executa la funció per a diversos continents i mostra els resultats obtinguts.

```python
def resumen_continente(df, continente):
    df_filtrado = df[df["Continent"] == continente] # filtrado por "continente"
    resumen = {
        "Regiones distintas": df_filtrado["Region"].nunique(), # número de regiones diferentes que hay dentro del continente
        "Número de países": df_filtrado["Name"].count(), # número de países dentro del continente
        "Superficie máxima": df_filtrado["SurfaceArea"].max(), # superficie máxima registrada en el continente
        "Superficie mínima": df_filtrado["SurfaceArea"].min(), # superficie mínima registrada
        "Población media": df_filtrado["Population"].mean(), # población media de los países del continente
        "Esperanza de vida promedio": df_filtrado["LifeExpectancy"].mean() # promedio de la esperanza de vida de los países del contiente
    }
    
    return resumen
```

L’objectiu és identificar quins països tenen menys espai disponible per habitant, fet que pot estar relacionat amb situacions de sobrepoblació, pressió urbana o desigualtat en l’ús del territori.

Per dur a terme aquest estudi, ens han proporcionat un conjunt de dades (Country(in).csv) que conté informació detallada sobre la població, la superfície i altres atributs de cada país.

Crea una funció en Python que rebi com a paràmetres:
1. Un DataFrame amb dades de països.
2. El nombre n de països que es volen obtenir com a resultat.

La funció ha de calcular, per a cada país, la quantitat de metres quadrats disponibles per persona, utilitzant la següent fórmula:

Metres per persona = SurfaceArea × 1.000.000 / Population

La funció ha de retornar els n països amb menys espai disponible per habitant.

Executa la funció per obtenir els 10 països amb menys espai per persona i després per obtenir-ne 20.

```python
import pandas as pd

df_country = pd.read_csv("Country(in).csv", sep=";", encoding="latin-1")
df_country
```

Crea una funció en Python que rebi com a paràmetres:
1. Un DataFrame amb dades de països.
2. El nombre n de països que es volen obtenir com a resultat.

La funció ha de calcular, per a cada país, la quantitat de metres quadrats disponibles per persona, utilitzant la següent fórmula:

Metres per persona = SurfaceArea × 1.000.000 / Population

La funció ha de retornar els n països amb menys espai disponible per habitant.

```python
def menos_espacio(df, n):
    df["Metros_por_persona"] = (df["SurfaceArea"] * 1_000_000) /df["Population"]
    return df.sort_values(by="Metros_por_persona").head(n)[["Name", "Metros_por_persona"]]
```

```python
menos_espacio(df_country, 10)
```

```python
menos_espacio(df_country, 20)
```

```python
def menos_espacio(df, n):
    df["Metros_por_persona"] = (df["SurfaceArea"] * 1_000_000) /df["Population"]
    return df.sort_values(["Metros_por_persona"], ascending=True).head(n)[["Name", "Metros_por_persona"]]

# aquí se usar sort_values sin el "by" y con ascending=True
```

```python
menos_espacio(df_country, 10)
```

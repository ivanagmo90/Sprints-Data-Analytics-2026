### 1.Calculadora de índice de masa corporal

```python
def calcular_imc():
    """
    Solicita el peso y la altura al usuario, valida los datos,
    calcula el IMC y clasifica el resultado en su categoría.
    """
    print("--- CALCULADORA DE IMC ---")

    # 1. Datos de entrada
    try:
        # Solicitamos los datos y los convertimos a decimales
        peso = float(input("Introduce tu peso en kilogramos (ej. 75,5): "))
        altura = float(input("Introduce tu altura en metros (ej. 1.75): "))
    except ValueError:
        # Si el usuario introduce letras o un formato incorrecto el programa no se cuelga
        return "Por favor, introduce sólo números válidos utilizando el punto (.) como separador decimal"
    
    # 2. Valores razonables (rangos)
    if peso <= 10 or peso >300:
        return "El peso introducido no parece un valor razonable"
    
    if altura <= 0.4 or altura > 2.5:
        return "La altura introducida no parece un valor razonble"
    
    # 3. Cálculo del IMC
    imc = round(peso / (altura ** 2), 2)

    # 4. Clasificación según OMS
    if imc < 18.5:
        categoria = "Bajo peso"
    elif 18.5 <= imc < 25.0:
        categoria = "Peso normal"
    elif 25.0 <= imc < 30.0:
        categoria = "Sobrepeso"
    else:
        categoria = "Obesidad"
    
    # 5. Retorno del resultado
    resultado = f"Tu ICM es {imc} y tu estado es {categoria}."
    return resultado
```

### 2. Conversor de temperaturas

```python
def conversor_temperaturas():
    """
    Solicita una unidad y valor. Devuelve la equivalencia de esa temperatura
    en el resto de unidades.
    """
    print("--- CONVERSOR DE TEMPERATURAS ----")
    print("Unidades: C (Celsius), F (Fahrenheit), K (Kelvin), RE (Réaumur), RA (Rankine)")

    # 1. Datos de entrada
    origen = input("Introduce la unidad de origen (C, F, K, RE, RA): ").strip().upper()
    unidades_validas = {"C", "F", "K", "RE", "RA"}
    if origen not in unidades_validas:
        return f"Error '{origen}' no es una unidad válida. Selecciona C, F, K, RE o RA."
    
    # 2. Valor numérico
    try:
        valor = float(input(f"Introduce el valor de la temperatura"))
    except ValueError:
        return "Por favor, introduce un número válida. Usa el punto (.) como decimal"
    
    # 3. Convertir la unidad de origen a la unidad puente (Celsius)
    a_celsius = {
        "C": lambda x: x,
        "F": lambda x: (x - 32) / 1.8,
        "K": lambda x: x - 273.15,
        "RE": lambda x: x * 1.25,
        "RA": lambda x: (x / 1.8) - 273.15
    }

    celsius_base = a_celsius[origen](valor)

    # cero absoluto (-273.15 °C)
    if celsius_base < -273.15:
        return "La temperatura introducida está por debajo del cero absoluto"
    
    # 4. Calcular y almacenar los números en el diccionario
    conversiones = {
        "C": round(celsius_base, 1),
        "F": round((celsius_base * 1.8) + 32, 1),
        "K": round(celsius_base + 273.15, 1),
        "RE": round(celsius_base * 0.8, 1),
        "RA": round((celsius_base + 273.15) * 1.8, 1)
    }

    # 5. Diccionario de símbolos y etiquetas
    simbolos = {"C": "°C", "F": "°F", "K": "°K", "RE": "°RE", "RA": "°RA"}
    
    # 6. Eliminamos la unidad de origen
    conversiones.pop(origen)

    # 7. Creamos la lista de textos formateados combinando clave-valor
    valores_restantes = ", ".join(
        f"{num}{simbolos[unidad]}" for unidad, num in conversiones.items()
    )

    # Retorno del resultado
    return f"{valor}{simbolos[origen]} equivale a {valores_restantes}."
```

### 3. Contador de palabras de un texto

```python
def contador_palabras():
    """
    Solicita un texto al usuario, limpia signos de puntuación,
    cuenta la frecuenca de cada palabra y calcula la longitud media.
    """
    print("--- Contador de palaras ---")
    texto_usuario = input("Introduce tu texto aquí: ").strip()

    # Validamos que el usuario no deje el campo vacío
    if not texto_usuario:
        return "El texto introducido está vacío"
    
    # 1. Limpieza
    # Pasamos todo a minúscualas
    texto_limpio = texto_usuario.lower()

    # Eliminamos signos de puntuación
    signos_puntuacion = [".", ",", ";", ":", "?", "!", "¿", "¡", "(", ")", '"', "-", "_"]
    for signo in signos_puntuacion:
        texto_limpio = texto_limpio.replace(signo, "")
    
    # 2. Separación en palabras
    palabras = texto_limpio.split()

    # verificación por si el texto sólo contenía signos de puntuación
    if len(palabras) == 0:
        return "El texto no contiene palabras válidas para analizar"
    
    # 3. Conteo de palabras
    conteo_palabras = {}
    for palabra in palabras:
        if palabra in conteo_palabras:
            conteo_palabras[palabra] += 1   #si ya existe sumamos 1
        else:
            conteo_palabras[palabra] = 1    #si es nueva, la registramos con 1

    #4. Cálculo longitud media
    # Sumamos las letras de todas las palabras y dividimos entre el total
    total_letras = sum(len(palabra) for palabra in palabras)
    longitud_media = round(total_letras / len(palabras), 2)

    # 5. Retorno
    resultado_final = {
        "Frecuencia de palabras": conteo_palabras,
        "Total palabras": len(palabras),
        "Longitud media de palabras": longitud_media
    }

    return resultado_final
```

### 4. Diccionario inverso (con posibilidad de duplicados)

```python
def diccionario_inverso(diccionario_cliente):
    """
    Invierte las claves y los valores de un diccionario
    """
    invertido = {}
    duplicados = []

    for clave, valor in diccionario_cliente.items():
        if valor not in invertido:
            invertido[valor] = clave

        else:
            if not isinstance(invertido[valor], list):  #pregunta si es un lista (FALSE), if not lo invierte (TRUE)
                invertido[valor] = [invertido[valor]]
            
            invertido[valor].append(clave)

            if valor not in duplicados:
                duplicados.append(valor)

    if duplicados:
        detalles = [f"'{val}': {invertido[val]}" for val in duplicados]
        mensaje = f"Elementos repetidos -> {', '.join(detalles)}"
    
    else:
        mensaje = "No se han encontrado valores duplciados"

    return invertido, mensaje
```

### 1. Contador y ordenador de palabras de un texto

```python
def leer_txt(ruta_archivo):
    with open(ruta_archivo, "r", encoding="utf-8") as archivo:
        return archivo.read()
    
def limpiar_contar_palabras(texto):
    texto_limpio = texto.lower()
    signos_puntuacion = [".", ",", ";", ":", "?", "!", "¿", "¡", "(", ")", '"', "-", "_"]
    for signo in signos_puntuacion:
        texto_limpio = texto_limpio.replace(signo, "")

    palabras = texto_limpio.split()

    conteo = {}
    for palabra in palabras:
        conteo[palabra] = conteo.get(palabra, 0) + 1
    return conteo

def agrupar_inicial(diccionario_conteo):
    agrupado = {}
    for palabra, frecuencia in diccionario_conteo.items():
        inicial = palabra[0]
        if inicial not in agrupado:
            agrupado[inicial] = {}
        agrupado[inicial][palabra] = frecuencia
    return agrupado
```

```python
def contar_ordenar_texto(ruta_archivo):
    texto_original = leer_txt(ruta_archivo)
    conteo_base = limpiar_contar_palabras(texto_original)
    diccionario_anidado = agrupar_inicial(conteo_base)

    # Para ordenar las palabras internas
    for inicial, diccionario_interno in diccionario_anidado.items():
        diccionario_anidado[inicial] = dict(sorted(diccionario_interno.items()))
    
    # Para ordenar las letras iniciales
    return dict(sorted(diccionario_anidado.items()))
```

### 2. Conversión de tipos de datos

```python
def conversion_datos(lista_sucia):
    floats = []
    no_floats = []

    # creamos una copia de la lista original para vaciarla de forma segura
    pila = list(lista_sucia)

    while pila:
        # extraemos el primer elemento de la lista
        elemento = pila.pop(0)

        # Si el elemento es otra lista/tupla/set, sus partes se añaden al inicio de la pila
        if isinstance(elemento, (list, tuple, set)):
            # insertamos los sub-elementos al principio para mantener el orden
            pila = list(elemento) + pila
        else:
            try:
                floats.append(float(elemento))
            except (ValueError, TypeError):
                no_floats.append(elemento)
    
    return floats, no_floats
```

### 1. Generador de contraseñas

```python
import numpy as np
import string
import pyperclip
```

```python
def obtener_bancos_caracteres(mayusculas, minusculas, numeros, signos):
    categorias = []

    if mayusculas:
        categorias.append(list(string.ascii_uppercase))
    if minusculas:
        categorias.append(list(string.ascii_lowercase))
    if numeros:
        categorias.append(list(string.digits))
    if signos:
        #se limitan los signos
        categorias.append([",", "-", "$", "?", "!", "@", "#", "%", "&", "*"])

    return categorias

def validar_parametros(longitud, categorias):
    if not categorias:
        raise ValueError("Error: Debes activar al menos un tipo de carácter.")
    
    if longitud < len(categorias):
        raise ValueError(f"Error: La longitud es muy corta. Necesitas al menos {len(categorias)} caracteres.")
    
def generar_caracteres_base(categorias):
    return [np.random.choice(banco) for banco in categorias]

def generar_caracteres_restantes(longitud, caracteres_base, categorias):
    longitud_restante = longitud - len(caracteres_base)

    #combinamos todos los bancos activos en una única lista plana
    todos_los_caracteres = []
    for banco in categorias:
        todos_los_caracteres.extend(banco)
    
    #seleccionamos el resto de caracteres necesarios del golpe con Numpy
    return list(np.random.choice(todos_los_caracteres, size=longitud_restante))

def mezclar_unir_cadena(lista_caracteres):
    np.random.shuffle(lista_caracteres)         #truco de ordenación in-place de NumPy
    return "".join(lista_caracteres)

def ejecutar_copiado_portapapeles(contrasena_texto):
    """
    Extra: Gestiona el copiado automático al sistema
    """

    try:
        pyperclip.copy(contrasena_texto)
        return "Contraseña generada con éxito y copiada a portapapeles (Ctrl+V)."
    except Exception:
        return "Contraseña generada con éxito, pero el entorno bloqueó el acceso al portapapeles."
```

```python
def crear_contrasena(longitud, mayusculas=True, minusculas=True, numeros=True, signos=False):
    # 1. construimos y validamos los bancos de caracteres necesarios
    categorias_activas = obtener_bancos_caracteres(mayusculas, minusculas, numeros, signos)
    validar_parametros(longitud, categorias=categorias_activas)

    # 2. generamos la base obligatoria y rellenamos el resto
    componentes_base = generar_caracteres_base(categorias_activas)
    componentes_restantes = generar_caracteres_restantes(longitud, componentes_base, categorias_activas)

    # 3. fusionamos los bloques de caracteres y los barajamos
    lista_completa = componentes_base + componentes_restantes
    contrasena_final = mezclar_unir_cadena(lista_completa)

    # 4. enviamos el resultado al portapapeles del sistema operativo
    mensaje_estado = ejecutar_copiado_portapapeles(contrasena_final)

    # 5. construcción del string unico final
    salida_completo = f"{mensaje_estado}\nResultado almacenado listo para el cliente: {contrasena_final}"

    return salida_completo
```

```python
mi_contrasena = crear_contrasena(10, True, True, True, True)
print(mi_contrasena)
```

```python
import requests

# URL base corregida y exacta para las peticiones
BASE_URL = "https://jsonplaceholder.typicode.com"

# Definición de los endpoints a consultar
endpoints = {
    "Publicaciones (/posts)": f"{BASE_URL}/posts",
    "Usuarios (/users)": f"{BASE_URL}/users",
    "Tareas (/todos)": f"{BASE_URL}/todos",
}

print("--- EJERCICIOS 1 Y 2: CONSULTAS Y CÓDIGOS DE ESTADO ---")
for nombre, url in endpoints.items():
    response = requests.get(url)
    codigo_estado = response.status_code

    if codigo_estado == 200:
        datos = response.json()
        total_recursos = len(datos)
        print(f"{nombre} -> Código de estado: {codigo_estado} | Cantidad total: {total_recursos}")
    else:
        print(f"{nombre} -> Error inesperado. Código: {codigo_estado}")
```

```python
print("\n--- EJERCICIO 3: PETICIÓN A UN RECURSO INEXISTENTE (ERROR 404) ---")
# Forzamos el error consultando la publicación ID 999999 (inexistente)
url_inexistente = f"{BASE_URL}/posts/999999"
response_404 = requests.get(url_inexistente)

print(f"URL consultada: {url_inexistente}")
print(f"Código de estado recibido de forma correcta: {response_404.status_code}")
```

```python
print("--- EJERCICIO 4: PETICIÓN POST (CREAR PUBLICACIÓN) ---")

url_post = f"{BASE_URL}/posts"

# Datos de la nueva publicación ficticia
nueva_publicacion = {
    "title": "Mi nueva publicación de prueba",
    "body": "Este es el contenido del cuerpo del mensaje para el Sprint 12.",
    "userId": 42,
}

# Realizamos la petición POST enviando los datos en el parámetro json
response_post = requests.post(url_post, json=nueva_publicacion)

print(f"Código de estado recibido: {response_post.status_code} (201 significa 'Created')")
print("Respuesta JSON devuelta por el servidor:")
print(response_post.json())
```

```python
print("--- EJERCICIO 5: PETICIÓN PATCH (MODIFICACIÓN PARCIAL) ---")

# Modificaremos únicamente el título de la publicación con ID 1
url_patch = f"{BASE_URL}/posts/1"

campos_a_modificar = {"title": "Este título ha sido modificado con un PATCH"}

# Realizamos la petición PATCH
response_patch = requests.patch(url_patch, json=campos_a_modificar)

print(f"Código de estado recibido: {response_patch.status_code}")
print("Respuesta JSON con el recurso modificado:")
print(response_patch.json())
```

```python
print("--- EJERCICIO 6: PETICIÓN DELETE (ELIMINAR RECURSO) ---")

# Eliminamos la publicación con ID 1
url_delete = f"{BASE_URL}/posts/1"

# Realizamos la petición DELETE
response_delete = requests.delete(url_delete)

print(f"Código de estado recibido: {response_delete.status_code}")
print("Respuesta JSON del servidor al eliminar:")
print(
    response_delete.json()
)  # JSONPlaceholder suele devolver un diccionario vacío {}
```
